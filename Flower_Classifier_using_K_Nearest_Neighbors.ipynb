**In this project, we will learn how to use the K-Nearest Neighbors (KNN) algorithm to classify the type of flower using a well-known dataset called the Iris dataset.**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import utils
```

```python
 #Load the Iris Dataset from sk-learn
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
print(iris.target_names)
```

```python
#Split the Data into Training and Testing Sets
from sklearn.model_selection import train_test_split
x = iris.data
y = iris.target
x_train, x_test, y_train, y_test =train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
x = iris.data[:, 2]
y = iris.data[:, 3]
labels = iris.target
plt.figure(figsize=(8, 6))
plt.scatter(x[labels == 0], y[labels == 0], color='red', label='setosa')
plt.scatter(x[labels == 1], y[labels == 1], color='green', label='versicolor')
plt.scatter(x[labels == 2], y[labels == 2], color='blue', label='virginica')

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Iris Dataset - Petal Length vs Width")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Train the KNN Model
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
model.fit(x_train, y_train)
```

```python
# Make Predictions
y_pred = model.predict(x_test)
```

```python
#Evaluate the Model Accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy",accuracy)
```

In this project, we used the K-Nearest Neighbors (KNN) algorithm to classify flowers using the Iris dataset. The model achieved 100% accuracy on the test set, showing excellent performance on this dataset.
